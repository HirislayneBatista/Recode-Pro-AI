# Importando Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importando o dataset 1

```python
# Importando o dataset
df_copa = pd.read_csv("/content/WorldCups.csv")
df_copa
```

```python
plt.figure(figsize=(20, 5))

# Total de titulos por seleção - grafico de barra
plt.subplot(1, 3, 1)
winners = df_copa['Winner'].value_counts() # Conta quantas vezes cada seleção venceu
winners.plot(kind='bar', color='orange')
plt.xlabel('País')
plt.ylabel('Número de Títulos')
plt.title('Total de Títulos por Seleção')

# Porcentagem de titulos por seleção - grafico de pizza
plt.subplot(1, 3, 2)
winners.plot(kind='pie', autopct='%1.1f%%', colormap='viridis')
plt.ylabel('') # Remove o rótulo do eixo y
plt.title('Porcentagem de Títulos por Pais')

# Gols por ano
plt.subplot(1, 3, 3)
goals = df_copa.groupby('Year')['GoalsScored'].sum()
goals.plot(kind='line', color='purple', marker='o', linestyle='-')
plt.xlabel('Ano')
plt.ylabel('Gols Marcados')
plt.title('Evolução dos Gols por Ano')
plt.grid()

# plt.plot(df_copa['Year'], df_copa['GoalsScored'], marker='o', linestyle='-',color='b')
# plt.xlabel('Ano')
# plt.ylabel('Gols Marcados')
# plt.title('Evolução dos gols nas Copas')
# plt.grid()

plt.show()
```

# Importando o dataset 2

```python
df_copa_jogos = pd.read_csv("/content/WorldCupPlayers.csv")
df_copa_jogos
```

```python
# prompt: Usando o DataFrame df_copa_jogadores: gere um garfico de team initials vs line-up
plt.figure(figsize=(20, 5))

# Contagem de jogadores por seleção
plt.subplot(1, 3, 1)
team_counts = df_copa_jogos["Team Initials"].value_counts().head(10)
# plt.bar(team_counts.index, team_counts.values)
team_counts.plot(kind='bar', color='orange')
plt.xticks(rotation=45)
plt.xlabel("Seleção")
plt.ylabel("Quantidade de Jogadores")
plt.title("Top 10 Seleções com Mais Jogadores")

# Contagem de jogadores por posição
plt.subplot(1, 3, 2)
df_copa_jogos['Position'].value_counts().plot(kind='bar', color='purple')
plt.xticks(rotation=45)
plt.xlabel("Posição")
plt.ylabel("Quantidade")
plt.title("Quantidade de Jogadores por Posição")

# Número de treinadores por país
plt.subplot(1, 3, 3)
df_copa_jogos['Coach Country'] = df_copa_jogos['Coach Name'].str.extract(r'\((.*?)\)')
df_copa_jogos['Coach Country'].value_counts().head(10).plot(kind='bar', color='cyan')
plt.title("Top 10 Países com Mais Treinadores")
plt.xticks(rotation=45)
plt.xlabel("País")
plt.ylabel("Quantidade")

plt.show()
```
