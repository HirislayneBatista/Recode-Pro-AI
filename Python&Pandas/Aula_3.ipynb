```python
import pandas as pd
```

# Uso do Merge com Pandas

```python
times = pd.DataFrame({
    'Time': ['Flamengo', 'Palmeiras', 'São Paulo', 'Corinthians'],
    'Cidade': ['Rio de Janeiro', 'São Paulo', 'São Paulo', 'São Paulo'],
    'Pontos': [59, 57, 52, 50]
})
 
tecnicos = pd.DataFrame({
    'Time': ['Flamengo', 'Palmeiras', 'São Paulo', 'Atlético-MG'],
    'Técnico': ['Tite', 'Abel Ferreira', 'Dorival Júnior', 'Cuca']
})
 
display(times)
display(tecnicos)
```

```python
result = pd.merge(times, tecnicos, on='Time', how='outer')
display(result)
```

2. Crie um DataFrame patrocinadores contendo os times e seus respectivos patrocinadores. Em seguida, faça um merge com o DataFrame times para adicionar a coluna de patrocinadores.

```python
patrocinadores = pd.DataFrame({
    'Time': ['Flamengo', 'Palmeiras', 'São Paulo', 'Atlético-MG'],
    'Patrocínio': ['Nike', 'Adidas', 'Spotify', 'Coca-Cola']
})

display(patrocinadores)
```
